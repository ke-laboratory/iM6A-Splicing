## Grab sequence

```python
import numpy as np
import pandas as pd
```

```python
Data = pd.read_csv("/home/morgenlefay/iM6A/Annotation/UniqmRNA_GENCODE_VM7com_mm10_RefSeqpriority.txt", sep="\t", header=None)
```

```python
Data.drop([0,11,13,14,15], axis=1, inplace=True)
```

```python
Data
```

```python
Data[1] = Data[12]
```

```python
Data.drop([12], axis=1, inplace=True)
```

```python
Data
```

```python
Data.columns = ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]
```

```python
Data
```

### Select coding genes

```python
idx = (Data["cdsStart"] != Data["cdsEnd"])
```

```python
Data = Data.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]]
```

```python
Data
```

### Keep standard chromosomes

```python
Data = Data[:22357]
```

### exonCount > 2

```python
Data = Data[Data.exonCount > 2]
```

```python
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["LastExonStart"] = 0
Data["LastExonEnd"] = 0
```

```python
Data
```

```python
for i in range(len(Data)):
    if Data.loc[i, "strand"] == "+":
        Data.loc[i,("LastExonStart")] = Data.loc[i, "exonStarts"].split(",")[-2]
        Data.loc[i,("LastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[-2]
    elif Data.loc[i, "strand"] == "-":
        Data.loc[i,("LastExonStart")] = Data.loc[i, "exonStarts"].split(",")[0]
        Data.loc[i,("LastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[0]
```

```python
Data
```

```python
for i in range(len(Data)):
    if Data.loc[i, "strand"] == "+":
        Data.loc[i,("SecondToLastExonStart")] = Data.loc[i, "exonStarts"].split(",")[-3]
        Data.loc[i,("SecondToLastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[-3]
    elif Data.loc[i, "strand"] == "-":
        Data.loc[i,("SecondToLastExonStart")] = Data.loc[i, "exonStarts"].split(",")[1]
        Data.loc[i,("SecondToLastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[1]
```

```python
Data
```

```python
StopCodon = pd.read_csv("/home/morgenlefay/iM6A/Annotation/StopCodon.csv")
```

```python
StopCodon
```

```python
StopCodon = StopCodon[["name","chrom","strand","txStart","txEnd"]]
```

```python
Data = pd.merge(Data, StopCodon, how="inner", on=["name","chrom","strand","txStart","txEnd"])
```

```python
Data
```

```python
Group = pd.read_csv("/home/morgenlefay/iM6A/Annotation/Group.csv")
```

```python
Group
```

```python
Group = Group[Group["Group"] != 1]
Group = Group[["name","chrom","strand","txStart","txEnd","Group"]]
Data = pd.merge(Data, Group, how="inner", on=["name","chrom","strand","txStart","txEnd"])
```

```python
Data
```

```python
Data.to_csv("Temp/MultiExon.csv", index=0)
```
