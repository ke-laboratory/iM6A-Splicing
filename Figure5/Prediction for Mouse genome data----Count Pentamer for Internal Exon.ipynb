### Motif

```python
from itertools import product
comb = list(product(["A","T","C","G"],repeat=5))
Mer = []
for i in range(len(comb)):
    value = "".join(comb[i])
    Mer.append(value)
Motif = dict(zip(Mer, [0]*len(Mer)))
```

```python
Data = Result
```

```python
Sequence = Data["Sequence"].tolist()
```

```python
for i in range(len(Data)):
    Seq = Sequence[i]
    Seq = Seq.upper()
    
    if len(Seq) <=105:
        for j in range(0, len(Seq)-5, 1):
            string = Seq[j:j+5]
            Motif[string] = Motif[string] + 1
    
    if len(Seq) > 105:
        Seq = Seq[0:105]
        for j in range(0, 105-5, 1):
            string = Seq[j:j+5]
            Motif[string] = Motif[string] + 1
```

```python
Count = pd.DataFrame([Motif])
```

```python
Count = Count.T
```

```python
Count = Count.reset_index(drop = False)
```

```python
Count.columns = ["Motif","C2"]
```
