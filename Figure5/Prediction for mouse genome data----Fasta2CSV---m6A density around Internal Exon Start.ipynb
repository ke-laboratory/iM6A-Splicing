```python
from Bio import SeqIO
from Bio.Seq import Seq
import numpy as np
import pandas as pd
import random
```

```python
meta=[]
sequence=[]
```

```python
seq = ('Temp/MultiExon.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Fasta
```

```python
Data = pd.read_csv("Temp/MultiExon.csv")
```

```python
Data
```

```python
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result
```

```python
Result.drop(["Meta"], axis=1, inplace=True)
```

```python
Result
```

```python
Result["GeneLength"] = Result["txEnd"] - Result["txStart"]
```

```python
Result
```

```python
Fasta = Result[["name","chrom","strand","Sequence"]]
```

```python
Data = Result.drop(["Sequence"], axis=1)
```

```python
Data 
```

```python
Data = Data.sort_values(by=["chrom","txStart"], axis=0, ascending=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data = Data[Data.exonCount >= 3]
Data = Data.reset_index(drop = True)
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive = Positive[(Positive["m6AStart"]>=Positive["exonStart"]) & (Positive["m6AEnd"]<=Positive["exonEnd"])]
Negative = Negative[(Negative["m6AStart"]>=Negative["exonStart"]) & (Negative["m6AEnd"]<=Negative["exonEnd"])]
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["M6APOS"] = (Positive["m6AStart"]-Positive["exonStart"]) + Positive["Start"]
Negative["M6APOS"] = (Negative["exonEnd"]-Negative["m6AEnd"]) + Negative["Start"]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```
