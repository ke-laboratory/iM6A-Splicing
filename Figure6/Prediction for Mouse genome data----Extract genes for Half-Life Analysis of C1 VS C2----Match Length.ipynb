```python
C2 = Data[Data["C1"]==0.0]
C2 = C2.reset_index(drop=True)
```

```python
C2
```

```python
cDNA = C1["cDNALength"].tolist()
```

```python
data2 = pd.DataFrame(columns = C2.columns.to_list())
```

```python
for i in range(len(cDNA)):
    
    length = cDNA[i]
    df = C2[(C2["cDNALength"]>=(length-25)) & (C2["cDNALength"]<=(length+25))]
    
    if len(df)>=1:
        df = df.sample(n=1, random_state=1)
        data2 = data2.append(df)
        C2 = C2.append(df)
        C2 = C2.drop_duplicates(subset=["name"], keep=False)
        C2 = C2.reset_index(drop = True)
        
    if len(df)==0:
        df = C1[C1["cDNALength"]==length]
        df = df.sample(n=1, random_state=1)
        C1 = C1.append(df)
        C1 = C1.drop_duplicates(subset=["name"], keep=False)
```
