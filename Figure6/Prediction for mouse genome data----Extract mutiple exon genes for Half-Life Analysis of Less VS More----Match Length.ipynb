```python
Data = Data[Data["Count"]>0]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data = Data[(Data["exonCount"]>=2)]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
data1 = Data[(Data["exonCount"]<=6)]
data1 = data1.reset_index(drop = True)
```

```python
data1
```

```python
cDNA = data1["cDNALength"].tolist()
```

```python
data2 = pd.DataFrame(columns = Data.columns.to_list())
```

```python
data2
```

```python
data3 = Data[Data["exonCount"]>6]
data3 = data3.reset_index(drop = True)
```

```python
data3
```

```python
for i in range(len(cDNA)):
    
    length = cDNA[i]
    df = data3[(data3["cDNALength"]>=(length-25)) & (data3["cDNALength"]<=(length+25))]
    
    if len(df)>=1:
        df = df.sample(n=1, random_state=1)
        data2 = data2.append(df)
        data3 = data3.append(df)
        data3 = data3.drop_duplicates(subset=["name"], keep=False)
        data3 = data3.reset_index(drop = True)
        
    if len(df)==0:
        df = data1[data1["cDNALength"]==length]
        df = df.sample(n=1, random_state=1)
        data1 = data1.append(df)
        data1 = data1.drop_duplicates(subset=["name"], keep=False)
```

```python
Data = data1.append(data2)
Data = Data.reset_index(drop = True)
```
