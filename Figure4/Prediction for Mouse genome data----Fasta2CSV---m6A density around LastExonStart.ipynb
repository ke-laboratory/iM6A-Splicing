```python
from Bio import SeqIO
from Bio.Seq import Seq
import numpy as np
import pandas as pd
import random
```

```python
Data = pd.read_csv("Temp/mm10.csv")
```

```python
Data
```

```python
Data["GeneLength"] = Data["txEnd"] - Data["txStart"]
```

```python
Data = Data.sort_values(by=["chrom","txStart"], axis=0, ascending=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Exon = pd.read_csv("/home/morgenlefay/iM6A/Chapter2/Figure1/MouseExonBed.csv")
```

```python
Exon
```

```python
M6A = pd.read_csv("iM6A_Mouse_LastIntronDeletion.csv")
```

```python
M6A
```

```python
M6A = M6A[["chrom","Start","End","name","strand","Deletion"]]
```

```python
M6A.columns = ["chrom","m6AStart","m6AEnd","name","strand","Probabilty"]
```

```python
Data = pd.merge(M6A, Data, how="inner", on=["chrom","name","strand"])
```

```python
Data
```

```python
Data = pd.merge(Data, Exon, how="inner", on=["chrom","name","strand"])
```

```python
Data = Data[(Data["m6AStart"]>=Data["Start"]) & (Data["m6AEnd"]<=Data["End"])]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["M6APOS"] = (Positive["m6AStart"]-Positive["Start"]) + (Positive["CumSum"]-Positive["ExonLength"])
Negative["M6APOS"] = (Negative["End"]-Negative["m6AEnd"]) + (Negative["CumSum"]-Negative["ExonLength"])
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.drop(["Start","End","ExonOrder","ExonLength","CumSum"], axis=1, inplace=True)
```

```python
Data
```

```python
Data = pd.merge(Data, Exon, how="inner", on=["chrom","name","strand"])
```

```python
Data
```

```python
Data = Data[(Data["Start"]==Data["LastExonStart"]) & (Data["End"]==Data["LastExonEnd"])]
```

```python
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["POS"] = Data["CumSum"] - Data["ExonLength"]
```

```python
Data
```

```python
Data.drop(["Start","End","ExonOrder","ExonLength","CumSum"], axis=1, inplace=True)
```

```python
Data
```

```python
Data["Distance"] = Data["M6APOS"] - Data["POS"]
```

```python
Data = Data[(Data["Distance"]>= -1000) & (Data["Distance"]<= 1000)]
```

```python
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.to_csv("Temp/mm10_M6A.csv", index=0)
```
