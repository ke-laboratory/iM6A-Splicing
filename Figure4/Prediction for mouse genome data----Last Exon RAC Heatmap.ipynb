```python
from Bio import SeqIO
from Bio.Seq import Seq
import numpy as np
import pandas as pd
import random
```

```python
meta=[]
sequence=[]
```

```python
seq = ('Temp/MultiExon.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Result["GeneLength"] = Result["txEnd"] - Result["txStart"]
Result["LastExonLength"] = Result["LastExonEnd"] - Result["LastExonStart"]
```

```python
Data = Data[Data["LastIntronLength"] > 5]
Data = Data.reset_index(drop = True)
```

### Heatmap

```python
Name = Data["name"].to_list()
Name = list(set(Name)) 
```

```python
Heatmap = pd.DataFrame({"Index":range(0, 200, 5)})
```

```python
for i in range(len(Name)):
    df = Data[Data["name"]==Name[i]]
    df = df.reset_index(drop = True)
    
    Position = df["m6APosition"].to_list()
    Prob = df["Probabilty"].to_list()
    Dict = dict(zip(Position,Prob))
    
    
    Value = []
    for j in range(40):
        Interval = [k for k in range(j*5,(j+1)*5,1)]
        
        List = []
        for num in Interval:
            if num in Dict.keys():
                List.append(1)
            if num not in Dict.keys():
                List.append(0)
        Value.append(sum(List)/5)
    
    df = pd.DataFrame({Name[i]:Value})
    Heatmap = pd.merge(Heatmap, df, left_index=True, right_index=True, how='outer')   
```
