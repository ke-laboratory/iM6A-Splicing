```python
import numpy as np
import pandas as pd
```

```python
Data = pd.read_csv("mm10_Fasta_LastIntronTruncation.csv")
```

```python
Data
```

```python
for i in range(len(Data)):
    Data.loc[i, "Donor"] = Data.loc[i,"Truncation"][:2].upper()
    Data.loc[i, "Acceptor"] = Data.loc[i,"Truncation"][-2:].upper()
```

```python
Data
```

```python
Data = Data[(Data["Donor"]=="GT") & (Data["Acceptor"]=="AG")]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.drop(["Donor","Acceptor"], axis=1, inplace=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.drop(["Truncation"], axis=1, inplace=True)
```

```python
Data = Data[Data["LastExonLength"]>=500]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Mutation = pd.read_csv("mm10_Fasta_Non_Donor_Acceptor.csv")
```

```python
Mutation
```

```python
Gene = Mutation[["name","chrom","strand"]]
```

```python
Gene = pd.merge(Data, Gene, how="inner", on=["name","chrom","strand"])
```

```python
Gene
```

```python
Data = Data.append(Gene)
Data.drop_duplicates(keep=False, inplace=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data = Data.sample(n=2326, random_state=1)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
for i in range(len(Data)):
    Data.loc[i,"Replace"] = Data.loc[i,"LastExonSequence"][100:200]
```

```python
Data
```

```python
Mutation["Replace"] = Data["Replace"]
```

```python
Mutation
```

```python
Mutation.to_csv("mm10_Fasta_Non_Donor_Acceptor_Replace.csv", index=0)
```
