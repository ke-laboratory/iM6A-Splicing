### Enhancer & Silencer

```python
Enhancer = pd.read_csv("Mouse_Motif_Regression_LastExon.csv")
```

```python
Enhancer = Enhancer.head(200)
```

```python
Enhancer = Enhancer["Name"].to_list()
```

```python
Silencer = pd.read_csv("Mouse_Motif_Regression_LastExon.csv")
```

```python
Silencer = Silencer.tail(200)
```

```python
Silencer = Silencer["Name"].to_list()
```

```python
motif_diFasta = dict()
positions = range(0, 210)
posi_str = []
for element in positions:
    posi_str.append(str(element))
motif_diFasta = dict.fromkeys(posi_str, 0)
```

```python
motif_dic2 = dict()
positions = range(0, 210)
posi_str = []
for element in positions:
    posi_str.append(str(element))
motif_dic2 = dict.fromkeys(posi_str, 0)
```

```python
for i in range(len(Fasta)):
    
    if Fasta.loc[i,"LastExonLength"]>=110:
        Seq = Fasta.loc[i,"Seq"][(Fasta.loc[i,"SecondToLastLength"]):(Fasta.loc[i,"SecondToLastLength"]+100)] + Fasta.loc[i,"Seq"][-(Fasta.loc[i,"LastExonLength"]):][0:110]
        
    if Fasta.loc[i,"LastExonLength"]<110:        
        Seq = Fasta.loc[i,"Seq"][(Fasta.loc[i,"SecondToLastLength"]):(Fasta.loc[i,"SecondToLastLength"]+100)] + Fasta.loc[i,"Seq"][-(Fasta.loc[i,"LastExonLength"]):]
        
    for j in range(0, len(Seq)-6, 1):
        string = Seq[j:j+5]
        if(string in Enhancer):
            motif_diFasta[str(j)]+=1
            
    for j in range(0, len(Seq)-6, 1):
        string = Seq[j:j+5]
        if(string in Silencer):
            motif_dic2[str(j)]+=1   
```

```python
Position = pd.DataFrame({"Index":range(0, 210)})
```

```python
Value=[]
for i in range(0,210,1):
    Value.append(int(motif_diFasta[str(i)]))
Position["Enhancer"]=Value
```

```python
Value=[]
for i in range(0,210,1):
    Value.append(int(motif_dic2[str(i)]))
Position["Silencer"]=Value
```

```python
df = pd.DataFrame({"Index":range(0, 210)})
for i in range(len(Fasta)):
    if Fasta.loc[i,"LastExonLength"]>=110:
        Seq = Fasta.loc[i,"Seq"][(Fasta.loc[i,"SecondToLastLength"]):(Fasta.loc[i,"SecondToLastLength"]+100)] + Fasta.loc[i,"Seq"][-(Fasta.loc[i,"LastExonLength"]):][0:110]
        
    if Fasta.loc[i,"LastExonLength"]<110:        
        Seq = Fasta.loc[i,"Seq"][(Fasta.loc[i,"SecondToLastLength"]):(Fasta.loc[i,"SecondToLastLength"]+100)] + Fasta.loc[i,"Seq"][-(Fasta.loc[i,"LastExonLength"]):]
        
    seq = Seq
    if len(seq) >= 210:
        lis = [1]*210
    if len(seq) < 210:
        lis = [1]*len(seq) + [0]*(210-len(seq))
        
    New = np.array(lis)
    New = pd.DataFrame({Fasta.loc[i,"name"]:New})
    df = pd.merge(df, New, left_index=True, right_index=True)
```

```python
df.drop(["Index"], axis=1, inplace=True)
df = df.T
```

```python
df.loc["col_sum"] = df.sum()
df = df.T
```

```python
Number = df[["col_sum"]]
```

```python
Number.columns = ["Number"]
```

```python
Data = pd.concat([Position, Number], axis=1)
```

```python
Data
```

```python
Data["Enhancer"] = Data["Enhancer"]/Data["Number"]
Data["Silencer"] = Data["Silencer"]/Data["Number"]
```
