### Read GenePred file

```python
import pandas as pd
```

```python
Data = pd.read_csv("genome.bed", sep="\t", header=None, error_bad_lines=False)
```

```python
Data
```

```python
Data.columns = ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]
```

```python
Data
```

### Coding genes

```python
idx = (Data["cdsStart"] != Data["cdsEnd"])
```

```python
Data = Data.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]]
```

```python
Data = Data.reset_index(drop=True)
```

```python
Data
```

### Gene name

```python
Table = pd.read_csv('gencode.vM7.annotation.gff3',sep='\t',header=None,skiprows=7)
```

```python
Table
```

```python
Table.drop(1,axis=1,inplace=True)
```

```python
Gene = Table.loc[Table[2] == 'transcript'].copy()
```

```python
Gene 
```

```python
Gene.loc[:,('GeneName')] = Gene.apply(lambda x:x[8].split(';')[6].split('=')[1],axis=1).values
Gene.loc[:,('name')] = Gene.apply(lambda x:x[8].split(';')[0].split('=')[1],axis=1).values
```

```python
Gene
```

```python
Gene.drop([0,2,3,4,5,6,7,8],axis=1,inplace=True)
```

```python
Gene
```

```python
Gene = Gene.reset_index(drop=True)
```

### Merge tables

```python
Merge = pd.merge(Data, Gene, how="inner", on = ["name"])
```

```python
Merge 
```

### Remove chrM

```python
idx = (Data["chrom"] != "chrM")
```

```python
Merge = Merge.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds", "GeneName"]]
```

```python
Merge = Merge.reset_index(drop=True)
```

```python
Merge
```

```python
Merge.to_csv("Gencode_vM7_mm10.csv", index=0)
```
