### Create Exon Bed for single

```python
Data = Single
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
for i in range(len(Positive)):
    df = pd.DataFrame(np.random.randn((Positive.loc[i,"exonCount"]), 3))
    df.columns = ["name", "chrom", "strand"]
    df["name"] = Positive.loc[i,"name"]
    df["chrom"] = Positive.loc[i,"chrom"]
    df["strand"] = Positive.loc[i,"strand"]
    
    exonStarts = Positive.loc[i, "exonStarts"].split(",")[0:-1]
    exonEnds = Positive.loc[i, "exonEnds"].split(",")[0:-1]
    
    Start = pd.DataFrame(exonStarts, columns=["Start"])
    End = pd.DataFrame(exonEnds, columns=["End"])
    
    df = pd.concat([df, Start], axis=1)
    df = pd.concat([df, End], axis=1)
    
    List = range(1, (Positive.loc[i,"exonCount"]+1))
    Count = pd.DataFrame(List, columns=["Count"])
    df = pd.concat([df, Count], axis=1)
    
    df["End"] = df["End"].astype(int)
    df["Start"] = df["Start"].astype(int)
    df["exonLength"] = df["End"] - df["Start"]
    df["cumsum"] = df["exonLength"].cumsum()
    
    df.to_csv("./Output3/{}.bed".format(Positive.loc[i,"name"]), sep="\t", index=False)
```

### Donor-Acceptor 

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Acceptor = dict.fromkeys(Positive["GeneName"].to_list())
```

```python
for key in Acceptor:
    Gene = key
    df = Positive[Positive["GeneName"]==Gene]
    df = df.reset_index(drop=True)
    
    List = []
    for i in range(len(df)):
        List.append((df.loc[i,"exonStart"], df.loc[i,"exonEnd"]))
    List = list(set(List))
    Dict = dict(zip(List, [0]*len(List)))
    
    for i in range(len(df)):
        Pair = (df.loc[i,"exonStart"],df.loc[i,"exonEnd"])
        
        if Pair in Dict.keys():
            Dict[Pair] = Dict[Pair] + 1
    
    Acceptor[key] = Dict
```

```python
Acceptor1 = dict.fromkeys(Negative["GeneName"].to_list())
```

```python
for key in Acceptor1:
    Gene = key
    df = Negative[Negative["GeneName"]==Gene]
    df = df.reset_index(drop=True)
    
    List = []
    for i in range(len(df)):
        List.append((df.loc[i,"exonStart"], df.loc[i,"exonEnd"]))
    List = list(set(List))
    Dict = dict(zip(List, [0]*len(List)))
    
    for i in range(len(df)):
        Pair = (df.loc[i,"exonStart"],df.loc[i,"exonEnd"])
        
        if Pair in Dict.keys():
            Dict[Pair] = Dict[Pair] + 1
    
    Acceptor1[key] = Dict
```
