### Fasta2CSV

```python
from Bio import SeqIO
import numpy as np
import pandas as pd
```

```python
meta=[]
sequence=[]
```

```python
seq = ('Temp/MultiExon.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Fasta
```

```python
Data = pd.read_csv("Temp/MultiExon.csv")
```

```python
Data
```

```python
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result
```

```python
Result.drop(["Meta"], axis=1, inplace=True)
```

```python
Result
```

```python
Result["Length"] = Result["txEnd"] - Result["txStart"]
```

```python
Result
```

```python
Data = Result
```

```python
Data 
```

```python
Data = Data.sort_values(by=["chrom","txStart"], axis=0, ascending=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["GeneLength"] = Data["txEnd"] - Data["txStart"]
Data["LastExonLength"] = Data["LastExonEnd"] - Data["LastExonStart"]
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["LastIntronLength"] = Positive["LastExonStart"] - Positive["SecondToLastExonEnd"]
Negative["LastIntronLength"] = Negative["SecondToLastExonStart"] - Negative["LastExonEnd"]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["Pre"] = Data["GeneLength"] - Data["LastExonLength"] - Data["LastIntronLength"]
```

```python
for i in range(len(Data)):
    Data.loc[i,("PreSequence")] = Data.loc[i,("Sequence")][:(Data.loc[i,("Pre")])]
```

```python
for i in range(len(Data)):
    Data.loc[i,("LastIntronSequence")] = Data.loc[i,("Sequence")][(Data.loc[i,("Pre")]):(Data.loc[i,("Pre")] + Data.loc[i,("LastIntronLength")])]
    Data.loc[i,("LastExonSequence")] = Data.loc[i,("Sequence")][(Data.loc[i,("Pre")] + Data.loc[i,("LastIntronLength")]):]
```

```python
Data.to_csv("Temp/mm10_Fasta.csv", index=0)
```
