```python
Data = pd.read_csv("/home/morgenlefay/iM6A/Annotation/UniqmRNA_GENCODE_VM7com_mm10_RefSeqpriority.txt", sep="\t", header=None)
Data.drop([0,11,13,14,15], axis=1, inplace=True)
Data[1] = Data[12]
Data.drop([12], axis=1, inplace=True)
Data.columns = ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]
idx = (Data["cdsStart"] != Data["cdsEnd"])
Data = Data.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]]
Data = Data[:22357]
Data = Data[Data.exonCount > 1]
Data = Data.reset_index(drop = True)
for i in range(len(Data)):
    if Data.loc[i, "strand"] == "+":
        Data.loc[i,("LastExonStart")] = Data.loc[i, "exonStarts"].split(",")[-2]
        Data.loc[i,("LastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[-2]
    elif Data.loc[i, "strand"] == "-":
        Data.loc[i,("LastExonStart")] = Data.loc[i, "exonStarts"].split(",")[0]
        Data.loc[i,("LastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[0]
for i in range(len(Data)):
    if Data.loc[i, "strand"] == "+":
        Data.loc[i,("SecondToLastExonStart")] = Data.loc[i, "exonStarts"].split(",")[-3]
        Data.loc[i,("SecondToLastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[-3]
    elif Data.loc[i, "strand"] == "-":
        Data.loc[i,("SecondToLastExonStart")] = Data.loc[i, "exonStarts"].split(",")[1]
        Data.loc[i,("SecondToLastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[1]
Data["SecondToLastExonStart"] = Data["SecondToLastExonStart"].astype(int)
Data["SecondToLastExonEnd"] = Data["SecondToLastExonEnd"].astype(int)
```

```python
Data = pd.merge(Data, M6A, how="inner", on=["chrom","name","strand"])
```

```python
Data = Data[(Data["Start"]>=Data["SecondToLastExonStart"]) & (Data["End"]<=Data["SecondToLastExonEnd"])]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["m6APosition"] = Positive["Start"] - Positive["SecondToLastExonStart"]
Negative["m6APosition"] = Negative["SecondToLastExonEnd"] - Negative["End"]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data["SecondToLastExonLength"] = Data["SecondToLastExonEnd"] - Data["SecondToLastExonStart"]
```
