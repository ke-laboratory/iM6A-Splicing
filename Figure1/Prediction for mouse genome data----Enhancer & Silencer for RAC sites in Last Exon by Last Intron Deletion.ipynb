```python
Data["GeneLength"] = Data["txEnd"] - Data["txStart"]
Data["LastExonLength"] = Data["LastExonEnd"] - Data["LastExonStart"]
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["LastIntronLength"] = Positive["LastExonStart"] - Positive["SecondToLastExonEnd"]
Negative["LastIntronLength"] = Negative["SecondToLastExonStart"] - Negative["LastExonEnd"]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data["Pre"] = Data["GeneLength"] - Data["LastExonLength"] - Data["LastIntronLength"]
```

```python
for i in range(len(Data)):
    Data.loc[i,("PreSequence")] = Data.loc[i,("Sequence")][:(Data.loc[i,("Pre")])]
```

```python
for i in range(len(Data)):
    Data.loc[i,("LastIntronSequence")] = Data.loc[i,("Sequence")][(Data.loc[i,("Pre")]):(Data.loc[i,("Pre")] + Data.loc[i,("LastIntronLength")])]
    Data.loc[i,("LastExonSequence")] = Data.loc[i,("Sequence")][(Data.loc[i,("Pre")] + Data.loc[i,("LastIntronLength")]):]
```

### Enhancer & Silencer

```python
Enhancer = pd.read_csv("Mouse_Motif_Regression_LastExon.csv")
```

```python
Enhancer = Enhancer.head(100)
```

```python
Enhancer = Enhancer["Name"].to_list()
```

```python
Silencer = pd.read_csv("Mouse_Motif_Regression_LastExon.csv")
```

```python
Silencer = Silencer.tail(100)
```

```python
Silencer = Silencer["Name"].to_list()
```

```python
motif_diFasta = dict()
positions = range(0, 110)
posi_str = []
for element in positions:
    posi_str.append(str(element))
motif_dic1 = dict.fromkeys(posi_str, 0)
```

```python
motif_dic2 = dict()
positions = range(0, 110)
posi_str = []
for element in positions:
    posi_str.append(str(element))
motif_dic2 = dict.fromkeys(posi_str, 0)
```

```python
for i in range(len(Data)):
    Seq1 = Data.loc[i,"Seq1"]
    Seq2 = Data.loc[i,"Seq2"]
    
    if len(Seq2)<59:
        Seq2 = Seq2 + "N"*(59-len(Seq2))

    Seq = Seq1 + Seq2
    Seq = Seq.upper()
   
    for j in range(0, len(Seq)-6, 1):
        string = Seq[j:j+5]
        if(string in Enhancer):
            motif_dic1[str(j)]+=1
            
    for j in range(0, len(Seq)-6, 1):
        string = Seq[j:j+5]
        if(string in Silencer):
            motif_dic2[str(j)]+=1         
```

```python
Position = pd.DataFrame({"Index":range(0, 110)})
```

```python
Value=[]
for i in range(0,110,1):
    Value.append(int(motif_dic1[str(i)]))
Position["Enhancer"]=Value
```

```python
Value=[]
for i in range(0,110,1):
    Value.append(int(motif_dic2[str(i)]))
Position["Silencer"]=Value
```

### Gene number

```python
df = pd.DataFrame({"Index":range(0, 110)})
```

```python
for i in range(len(Data)):
    
    Seq1 = Data.loc[i,"Seq1"]
    Seq2 = Data.loc[i,"Seq2"]
    
    Seq1_Length = Data.loc[i,"Seq1_Length"]
    Seq2_Length = len(Data.loc[i,"Seq2"])
    
    if Seq1_Length < 51:
        List1 = [0]*(51-Seq1_Length) + [1]*Seq1_Length
    if Seq1_Length >= 51:
        List1 = [1]*51
        
    if Seq2_Length < 59:
        List2 = [1]*Seq2_Length + [0]*(59-Seq2_Length)
    if Seq2_Length >= 59:
        List2 = [1]*59    

    lis = List1 + List2
    New = np.array(lis)
    New = pd.DataFrame({i:New})
    df = pd.merge(df, New, left_index=True, right_index=True)            
```

```python
df
```

```python
df.drop(["Index"], axis=1, inplace=True)
df = df.T
df.loc["col_sum"] = df.sum()
df = df.T
Number = df[["col_sum"]]
Number.columns = ["Number"]
```

```python
Number
```

```python
Data = pd.concat([Position, Number], axis=1)
```

```python
Data["Enhancer"] = Data["Enhancer"]/Data["Number"]
Data["Silencer"] = Data["Silencer"]/Data["Number"]
```
