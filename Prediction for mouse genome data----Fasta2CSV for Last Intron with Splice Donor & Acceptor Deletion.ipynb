```python
import numpy as np
import pandas as pd
```

```python
Data = pd.read_csv("mm10_Fasta_LastIntronTruncation.csv")
```

```python
Data
```

```python
for i in range(len(Data)):
    Data.loc[i, "Donor"] = Data.loc[i,"Truncation"][:2].upper()
    Data.loc[i, "Acceptor"] = Data.loc[i,"Truncation"][-2:].upper()
```

```python
Data
```

```python
Data = Data[(Data["Donor"]=="GT") & (Data["Acceptor"]=="AG")]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.drop(["Donor","Acceptor"], axis=1, inplace=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
for i in range(len(Data)):
    Data.loc[i,"Seq"] = Data.loc[i,"SecondToLastExonSequence"] + "CA" +Data.loc[i,"Truncation"][2:-2] + "TC" + Data.loc[i,"LastExonSequence"]
```

```python
Data
```

```python
Data.to_csv("mm10_Fasta1.csv", index=0)
```

### SD mutation

```python
Data = SD
```

```python
List = Data["name"].to_list()
List = list(set(List)) 
```

```python
for i in range(len(List)):
    df = Data[Data["name"]==List[i]]
    df = df.reset_index(drop = True)
    
    Position = df["Position"].tolist()
    Seq = df.loc[0,"Seq"]
    
    for j in Position:
        Seq = Seq[:(j+1)] + "CA" + Seq[(j+1+2):]
    
    df.drop(["Position"], axis=1, inplace=True)
    df = df.head(1)
    df.loc[0,"Seq"] = Seq
    df.to_csv("./SD/{}.bed".format(List[i]), sep="\t", index=False)
```

```python
path = "SD"
file = glob.glob(os.path.join(path, "*.bed"))
d = []
for f in file:
    d.append(pd.read_table(f))
data = pd.concat(d)
data = data.reset_index(drop = True)
```
