## Grab sequence

```python
import numpy as np
import pandas as pd
```

```python
Data = pd.read_csv("/home/morgenlefay/iM6A/Annotation/UniqmRNA_GENCODE_V19com_hg19_RefSeqpriority.txt", sep="\t", header=None)
```

```python
Data.drop([0,11,13,14,15], axis=1, inplace=True)
```

```python
Data
```

```python
Data[1] = Data[12]
```

```python
Data.drop([12], axis=1, inplace=True)
```

```python
Data
```

```python
Data.columns = ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]
```

```python
Data
```

### Select coding genes

```python
idx = (Data["cdsStart"] != Data["cdsEnd"])
```

```python
Data = Data.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]]
```

```python
Data
```

### Keep standard chromosomes

```python
Data = Data[:20536]
```

### exonCount >= 3

```python
Data = Data[Data.exonCount >= 3]
```

```python
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
StopCodon = pd.read_csv("/home/morgenlefay/iM6A/Annotation/HumanStopCodon.csv")
```

```python
StopCodon
```

```python
StopCodon = StopCodon[["name","chrom","strand","txStart","txEnd"]]
```

```python
Data = pd.merge(Data, StopCodon, how="inner", on=["name","chrom","strand","txStart","txEnd"])
```

```python
Data
```

```python
Group = pd.read_csv("/home/morgenlefay/iM6A/Annotation/HumanGroup.csv")
```

```python
Group
```

```python
Group = Group[Group["Group"] != 1]
Group = Group[["name","chrom","strand","txStart","txEnd","Group"]]
Data = pd.merge(Data, Group, how="inner", on=["name","chrom","strand","txStart","txEnd"])
```

```python
Data
```

```python
Data.to_csv("Temp/MultiExon.csv", index=0)
```

### Create Bed

```python
Bed = Data.iloc[:, [1, 3, 4, 0, 7, 2]]
```

```python
Bed.to_csv("Temp/MultiExon.bed", sep="\t", header=False, index=False)
```
