### Select coding genes

```python
idx = (Data["cdsStart"] != Data["cdsEnd"])
```

```python
Data = Data.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]]
```

```python
Data
```

### Keep standard chromosomes

```python
Data = Data[:20536]
```

```python
Data = Data[Data["chrom"] != "chrM"]
```

```python
Data = Data[Data.exonCount>1]
```

```python
Data
```

```python
Group = pd.read_csv("/home/morgenlefay/iM6A/Annotation/HumanGroup.csv")
```

```python
Group = Group[Group["Group"]!=1]
```

```python
Group = Group[["name","chrom","strand","txStart","txEnd","LastExonStart","LastExonEnd"]]
Data = pd.merge(Data, Group, how="inner", on=["name","chrom","strand","txStart","txEnd"])
```

```python
Data
```

```python
Data = Data.drop_duplicates(subset=["name"], keep=False)
Data = Data.reset_index(drop=True)
```

```python
Data
```

```python
Data.to_csv("Temp/hg19.csv", index=0)
```
