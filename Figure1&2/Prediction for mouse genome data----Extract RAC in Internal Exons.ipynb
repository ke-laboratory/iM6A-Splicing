```python
from Bio import SeqIO
from Bio.Seq import Seq
import numpy as np
import pandas as pd
import random
```

```python
meta=[]
sequence=[]
```

```python
seq = ('Temp/MultiExon.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Fasta
```

```python
Data = pd.read_csv("Temp/MultiExon.csv")
```

```python
Data
```

```python
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result
```

```python
Result.drop(["Meta","Sequence"], axis=1, inplace=True)
```

```python
Result
```

```python
Data = Result
```

```python
LongInternalExon = LongInternalExon.reset_index(drop = False)
LongInternalExon["index"] = LongInternalExon["index"].astype(str)
LongInternalExon["ID"] = "ID"+ LongInternalExon["index"]
LongInternalExon.drop(["index"], axis=1,inplace=True)
```

```python
M6A = pd.read_csv("/home/morgenlefay/iM6A/Chapter3/Figure1/iM6A_Mouse_IL.csv")
```

```python
M6A
```

```python
M6A.columns = ["chrom","m6AStart","m6AEnd","name","strand","WT","IL"]
```

```python
Data = pd.merge(M6A, Data, how="inner", on=["chrom","name","strand"])
```

```python
Data
```

```python
Data = pd.merge(Data, LongInternalExon, how="inner", on=["chrom","name","strand"])
```

```python
Data
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive = Positive[(Positive["m6AStart"]>=Positive["exonStart"]) & (Positive["m6AEnd"]<=Positive["exonEnd"])]
Negative = Negative[(Negative["m6AStart"]>=Negative["exonStart"]) & (Negative["m6AEnd"]<=Negative["exonEnd"])]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["m6APosition"] = Positive["m6AStart"] - Positive["exonStart"]
Negative["m6APosition"] = Negative["exonEnd"] - Negative["m6AEnd"]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```
