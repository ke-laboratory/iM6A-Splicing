### Get position for synonymous codon

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
```

```python
Gene = list(set(Positive["name"].to_list()))
```

```python
for i in range(len(Gene)):
    
    df = Positive[Positive["name"]==Gene[i]]
    df = df.sort_values(by=["ExonOrder"], ascending=["True"])
    df = df.reset_index(drop = True)
    
    Sequence = df["Sequence"].to_list()
    Sequence = "".join(Sequence)
    Sequence = Sequence.upper()
    
    Position = []
    for j in range(len(df)):
        List = list(range(df.loc[j,"Start"], df.loc[j,"End"], 1))
        Position = Position + List
    
    Deg = []
    if len(Sequence)%3 == 0: 
        for k in range(0, len(Sequence), 3): 
            Codon = Sequence[k:(k+3)]    
            
            if (Codon in ["TTT","TTC"]):
                Deg.append(Position[(k+2)])            
            if (Codon in ["ATT","ATC","ATA"]):
                Deg.append(Position[(k+2)])        
            if (Codon in ["GTT","GTC","GTA","GTG"]):    
                Deg.append(Position[(k+2)])      
            if (Codon in ["TCT","TCC","TCA","TCG"]):    
                Deg.append(Position[(k+2)])        
            if (Codon in ["CCT","CCC","CCA","CCG"]):    
                Deg.append(Position[(k+2)])           
            if (Codon in ["ACT","ACC","ACA","ACG"]):    
                Deg.append(Position[(k+2)])       
            if (Codon in ["GCT","GCC","GCA","GCG"]):    
                Deg.append(Position[(k+2)])         
            if (Codon in ["TAT","TAC"]):
                Deg.append(Position[(k+2)])               
            if (Codon in ["CAT","CAC"]):
                Deg.append(Position[(k+2)])        
            if (Codon in ["CAA","CAG"]):
                Deg.append(Position[(k+2)])        
            if (Codon in ["AAT","AAC"]):
                Deg.append(Position[(k+2)])           
            if (Codon in ["AAA","AAG"]):
                Deg.append(Position[(k+2)])    
            if (Codon in ["GAT","GAC"]):
                Deg.append(Position[(k+2)])      
            if (Codon in ["GAA","GAG"]):
                Deg.append(Position[(k+2)])      
            if (Codon in ["TGT","TGC"]):
                Deg.append(Position[(k+2)])    
            if (Codon in ["AGT","AGC"]):
                Deg.append(Position[(k+2)])                   
            if (Codon in ["GGT","GGC","GGA","GGG"]):    
                Deg.append(Position[(k+2)])               
            if (Codon in ["TTA","TTG","CTT","CTC","CTA","CTG"]):
                Deg.append(Position[(k)])                   
                Deg.append(Position[(k+2)])                     
            if (Codon in ["CGT","CGC","CGA","CGG","AGA","AGG"]):
                Deg.append(Position[(k)])                      
                Deg.append(Position[(k+2)])
            if (Codon in ["TAA","TAG","TGA"]):
                Deg.append(Position[(k+1)])                   
                Deg.append(Position[(k+2)])
    
    # Export
    data = pd.DataFrame({"POS":Deg})
    data["chrom"] = df.loc[0,"chrom"]
    data["name"] = df.loc[0,"name"]
    data["strand"] = df.loc[0,"strand"]
    data = data[["chrom","name","strand","POS"]]
    data.to_csv("./Output1/{}.bed".format(Gene[i]), sep="\t", index=False)    
```
