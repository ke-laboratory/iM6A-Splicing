### Create Exon Bed

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
for i in range(len(Positive)):
    df = pd.DataFrame(np.random.randn((Positive.loc[i,"exonCount"]), 3))
    df.columns = ["name", "chrom", "strand"]
    df["name"] = Positive.loc[i,"name"]
    df["chrom"] = Positive.loc[i,"chrom"]
    df["strand"] = Positive.loc[i,"strand"]
    
    exonStarts = Positive.loc[i, "exonStarts"].split(",")[0:-1]
    exonEnds = Positive.loc[i, "exonEnds"].split(",")[0:-1]
    exonInterval = Positive.loc[i, "Interval"].split(",")[0:-1]
    
    Start = pd.DataFrame(exonStarts, columns=["Start"])
    End = pd.DataFrame(exonEnds, columns=["End"])
    Interval = pd.DataFrame(exonInterval, columns=["Interval"])
    
    df = pd.concat([df, Start], axis=1)
    df = pd.concat([df, End], axis=1)
    df = pd.concat([df, Interval], axis=1)
    
    list = range(1, (Positive.loc[i,"exonCount"]+1))
    Count = pd.DataFrame(list, columns=["Count"])
    df = pd.concat([df, Count], axis=1)
    
    df["End"] = df["End"].astype(int)
    df["Start"] = df["Start"].astype(int)
    df["exonLength"] = df["End"] - df["Start"]
    df["cumsum"] = df["exonLength"].cumsum()
    
    df.to_csv("./Output1/{}.bed".format(Positive.loc[i,"name"]), sep="\t", index=False)
```

```python
for i in range(len(Negative)):
    df = pd.DataFrame(np.random.randn((Negative.loc[i,"exonCount"]), 3))
    df.columns = ["name", "chrom", "strand"]
    df["name"] = Negative.loc[i,"name"]
    df["chrom"] = Negative.loc[i,"chrom"]
    df["strand"] = Negative.loc[i,"strand"]
    
    exonStarts = Negative.loc[i, "exonStarts"].split(",")[0:-1]
    exonEnds = Negative.loc[i, "exonEnds"].split(",")[0:-1]
    exonInterval = Negative.loc[i, "Interval"].split(",")[0:-1]
    
    Start = pd.DataFrame(exonStarts, columns=["Start"])
    End = pd.DataFrame(exonEnds, columns=["End"])
    Interval = pd.DataFrame(exonInterval, columns=["Interval"])
    
    df = pd.concat([df, Start], axis=1)
    df = pd.concat([df, End], axis=1)
    df = pd.concat([df, Interval], axis=1)
    
    df.sort_index(inplace=True, ascending=False)
    df = df.reset_index(drop = True)
    
    list = range(1, (Negative.loc[i,"exonCount"]+1))
    Count = pd.DataFrame(list, columns=["Count"])
    df = pd.concat([df, Count], axis=1)
    
    
    df["End"] = df["End"].astype(int)
    df["Start"] = df["Start"].astype(int)
    df["exonLength"] = df["End"] - df["Start"]
    df["cumsum"] = df["exonLength"].cumsum()
    
    df.to_csv("./Output1/{}.bed".format(Negative.loc[i,"name"]), sep="\t", index=False)
```

### Enhancer & Silencer

```python
Enhancer = pd.read_csv("Mouse_Motif_Regression_LastExon.csv")
```

```python
Enhancer = Enhancer.head(100)
```

```python
Enhancer = Enhancer["Name"].to_list()
```

```python
Silencer = pd.read_csv("Mouse_Motif_Regression_LastExon.csv")
```

```python
Silencer = Silencer.tail(100)
```

```python
Silencer = Silencer["Name"].to_list()
```

```python
motif_diFasta = dict()
positions = range(0, 110)
posi_str = []
for element in positions:
    posi_str.append(str(element))
motif_dic1 = dict.fromkeys(posi_str, 0)
```

```python
motif_dic2 = dict()
positions = range(0, 110)
posi_str = []
for element in positions:
    posi_str.append(str(element))
motif_dic2 = dict.fromkeys(posi_str, 0)
```

```python
for i in range(len(Data)):
    Seq1 = Data.loc[i,"Seq1"]
    Seq2 = Data.loc[i,"Seq2"]
    
    if len(Seq1)<51:
        Seq1 = "N"*(51-len(Seq1)) + Seq1
    if len(Seq2)<59:
        Seq2 = Seq2 + "N"*(59-len(Seq2))

    Seq = Seq1 + Seq2
    Seq = Seq.upper()
   
    for j in range(0, len(Seq)-6, 1):
        string = Seq[j:j+5]
        if(string in Enhancer):
            motif_dic1[str(j)]+=1
            
    for j in range(0, len(Seq)-6, 1):
        string = Seq[j:j+5]
        if(string in Silencer):
            motif_dic2[str(j)]+=1      
```

```python
Position = pd.DataFrame({"Index":range(0, 110)})
```

```python
Value=[]
for i in range(0,110,1):
    Value.append(int(motif_dic1[str(i)]))
Position["Enhancer"]=Value
```

```python
Value=[]
for i in range(0,110,1):
    Value.append(int(motif_dic2[str(i)]))
Position["Silencer"]=Value
```

### Gene number

```python
df = pd.DataFrame({"Index":range(0, 110)})
```

```python
for i in range(len(Data)):
    
    Seq1_Length = Data.loc[i,"Seq1_Length"]
    Seq2_Length = Data.loc[i,"Seq2_Length"]
    
    if Seq1_Length < 51:
        List1 = [0]*(51-Seq1_Length) + [1]*Seq1_Length
    if Seq1_Length >= 51:
        List1 = [1]*51
        
    if Seq2_Length < 59:
        List2 = [1]*Seq2_Length + [0]*(59-Seq2_Length)
    if Seq2_Length >= 59:
        List2 = [1]*59    

    lis = List1 + List2
    New = np.array(lis)
    New = pd.DataFrame({i:New})
    df = pd.merge(df, New, left_index=True, right_index=True)              
```
