```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqUtils import nt_search
import numpy as np
import pandas as pd
import random
```

```python
meta=[]
sequence=[]
```

```python
seq = ('Temp/MultiExon.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Fasta
```

```python
Data = pd.read_csv("Temp/MultiExon.csv")
```

```python
Data
```

```python
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result
```

```python
Result.drop(["Meta"], axis=1, inplace=True)
```

```python
Result
```

```python
Result["GeneLength"] = Result["txEnd"] - Result["txStart"]
Result["LastExonLength"] = Result["LastExonEnd"] - Result["LastExonStart"]
```

```python
Result
```

```python
Data = Result
```

```python
Positive = Data[Data["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Data[Data["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive["LastIntronLength"] = Positive["LastExonStart"] - Positive["SecondToLastExonEnd"]
Negative["LastIntronLength"] = Negative["SecondToLastExonStart"] - Negative["LastExonEnd"]
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```

```python
Data["SecondToLastLength"] = Data["SecondToLastExonEnd"] - Data["SecondToLastExonStart"]
```

```python
Data
```

```python
Data = Data[Data["LastIntronLength"] >= 200]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["Pre"] = Data["GeneLength"] - Data["LastExonLength"] - Data["LastIntronLength"] - Data["SecondToLastLength"]
```

```python
Data
```

### Get sequence for Intron

```python
for i in range(len(Data)):
    Data.loc[i,("PreSequence")] = Data.loc[i,("Sequence")][:(Data.loc[i,("Pre")])]
```

```python
for i in range(len(Data)):
    Data.loc[i,("SecondToLastExonSequence")] = Data.loc[i,("Sequence")][(Data.loc[i,("Pre")]):(Data.loc[i,("Pre")] + Data.loc[i,("SecondToLastLength")])]    
    Data.loc[i,("LastIntronSequence")] = Data.loc[i,("Sequence")][(Data.loc[i,("Pre")] + Data.loc[i,("SecondToLastLength")]):(Data.loc[i,("Pre")] + Data.loc[i,("SecondToLastLength")] + Data.loc[i,("LastIntronLength")])]
    Data.loc[i,("LastExonSequence")] = Data.loc[i,("Sequence")][(Data.loc[i,("Pre")] + Data.loc[i,("SecondToLastLength")] + Data.loc[i,("LastIntronLength")]):]
```

```python
Data
```

```python
for i in range(len(Data)):
    Data.loc[i,("Truncation")] = (Data.loc[i,("LastIntronSequence")][0:100] + Data.loc[i,("LastIntronSequence")][-100:]).upper()
```

```python
Data
```

```python
Data.to_csv("mm10_Fasta_LastIntronTruncation.csv", index=0)
```
